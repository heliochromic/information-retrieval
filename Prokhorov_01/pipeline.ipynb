# Практичне 1. Словник #
## Імпортування всіх потрібних бібліотек та функцій з `functions.py` ##

```python
import os
from functions import calculate_frequency, to_csv, to_dict, serialize, deserialize, stats
```

## Будування словника ##
Отримання списку імен файлів у каталозі 'texts'

```python
FILENAMES = os.listdir('./texts')
FILENAMES
```

Виклик функції `calculate_frequency` щоб побудувати словник по заданому списку файлів та отримати його в відсортованому вигляді

```python
DICTIONARY = calculate_frequency(FILENAMES)
DICTIONARY
```

## Зберігання всіх термінів на диск ##
Зберігання в .csv формат

```python
to_csv(FILENAMES, DICTIONARY)
to_dict()
```

Зберігання сереалізованого об'єкту в .pkl форматі 

```python
serialize(DICTIONARY)
deserialize()
```

## Порівняння двох збережених форматів ##

```python
stats('output.csv', to_dict())
stats('serialized.pkl', deserialize())
```
